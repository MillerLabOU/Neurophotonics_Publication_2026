## Renamed 'frame_0199.png'  ->  'S14_031_frame_0199.png'

```python
import os

# --- Configuration ---
# 1. IMPORTANT: Set the path to the folder containing your images.
#    - For Windows, it might look like: r'C:\Users\YourUser\Desktop\MyImages'
#    - For macOS or Linux, it might look like: '/Users/YourUser/Desktop/MyImages'
folder_path = r"D:\Labeling\2025.5.12 Labeled Frames Anitah 104 Frames Total_renamed\Frames\Labeled frames_Anitah\S14_031 png lsci"

# 2. Define the prefix to add.
prefix = 'S14_031_'
# -------------------

try:
    # Get a list of all files in the directory
    all_files = os.listdir(folder_path)
    
    print(f"Found {len(all_files)} files. Starting renaming process...\n")

    # Loop through each file in the folder
    for filename in all_files:
        # Check to avoid re-naming files that already have the prefix
        if not filename.startswith(prefix):
            # Construct the old and new file paths
            old_file_path = os.path.join(folder_path, filename)
            new_file_name = f"{prefix}{filename}"
            new_file_path = os.path.join(folder_path, new_file_name)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f"Renamed '{filename}'  ->  '{new_file_name}'")

    print("\n✅ Renaming complete!")

except FileNotFoundError:
    print(f"❌ ERROR: The folder path '{folder_path}' was not found.")
    print("Please make sure the path is correct and try again.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Save averaged frames in a new folder 

```python
import os
import cv2
import numpy as np
from tqdm import tqdm # A library for creating smart progress bars

# --- Configuration ---
# 1. The folder where your original frames are located.
input_folder = r"D:\Labeling\2025.1.9 Dr Miller Video and Frames\Frames_LSCI\B7_004_sc"

# 2. The folder where the final averaged image will be saved.
output_folder = r"D:\Labeling\2025.10.8 HD LSCI images before and after averaging\After averaging\2025.4.28 Labeled Frames Jacqueline 9 Frames Total_renamed"

# 3. Define the range of frames to average (inclusive).
start_frame = 300
end_frame = 310
# --- End of Configuration ---

print("Starting frame averaging process...")

# Ensure the output directory exists, create it if it doesn't
os.makedirs(output_folder, exist_ok=True)
print(f"Output will be saved to: {output_folder}")

# --- Main Averaging Logic ---
sum_of_frames = None
frames_processed = 0
frame_height, frame_width, layers = 0, 0, 0

# Create a list of frame numbers to process
frame_numbers_to_process = range(start_frame, end_frame + 1)

# Loop through the specified frame numbers with a progress bar
for i in tqdm(frame_numbers_to_process, desc="Averaging Frames"):
    # Format the filename, e.g., 'frame_0000.png', 'frame_0001.png', etc.
    filename = f"frame_{i:04d}.png"
    file_path = os.path.join(input_folder, filename)

    if os.path.exists(file_path):
        # Read the image
        img = cv2.imread(file_path)
        
        if img is not None:
            # If this is the first image, initialize the numpy array for the sum
            if sum_of_frames is None:
                frame_height, frame_width, layers = img.shape
                # Use a 64-bit float to prevent overflow when summing pixel values
                sum_of_frames = np.zeros((frame_height, frame_width, layers), np.float64)

            # Add the current image's pixel values to the running sum
            sum_of_frames += img.astype(np.float64)
            frames_processed += 1
        else:
            print(f"Warning: Could not read image file: {filename}")
    else:
        print(f"Warning: File not found, skipping: {filename}")

# --- Final Calculation and Saving ---
if frames_processed > 0:
    # Calculate the average by dividing the sum by the number of frames
    averaged_frame_float = sum_of_frames / frames_processed

    # Convert the averaged frame back to an 8-bit integer format for saving as a PNG
    averaged_frame_uint8 = averaged_frame_float.astype(np.uint8)

    # Define the output filename
    output_filename = f"averaged_frame_{start_frame:04d}-{end_frame:04d}.png"
    output_path = os.path.join(output_folder, output_filename)

    # Save the final image
    cv2.imwrite(output_path, averaged_frame_uint8)
    
    print("\n✅ Success! Averaging complete.")
    print(f"Processed {frames_processed} frames.")
    print(f"Result saved as: {output_path}")
else:
    print("\n❌ Error: No frames were found or processed in the specified directory.")
```
